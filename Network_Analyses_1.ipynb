## 1. Define:

	(a) Subgraph
	
	 Um subgrafo de um grafo é um conjunto de vértices e arestas que pertencem ao grafo original. Ele pode ser obtido removendo alguns vértices e/ou 	arestas do grafo original.
	
	(b) Bipartite graph.
	
	 Um grafo bipartido é um tipo de grafo em que os vértices podem ser divididos em dois conjuntos distintos, de modo que todas as arestas conectam 	um vértice de um conjunto com um vértice do outro conjunto. Em outras palavras, não há arestas entre vértices do mesmo conjunto.
	
	(c) Hamiltonian graph
	
	 Um grafo hamiltoniano é um grafo que contém um caminho hamiltoniano, ou seja, um caminho que passa por todos os vértices do grafo exatamente uma 	   vez. Um ciclo hamiltoniano é um caminho hamiltoniano que forma um ciclo fechado.
	 
	(d) Eulerian graph.
	
	 Um grafo euleriano é um grafo que contém um caminho euleriano, ou seja, um caminho que passa por todas as arestas do grafo exatamente uma vez. Um         ciclo euleriano é um caminho euleriano que forma um ciclo fechado. Um grafo euleriano deve ter todos os seus vértices com grau par.
	

## 2. Describe how a breadth-first search algorithm works.

   É um método para percorrer um grafo, começa pelo vértice raiz e explora todos os vértices adjacentes ao mesmo nível antes de se mover para o próximo nível.Esse processo continua até que a fila esteja vazia, o que significa que todos os vértices alcançáveis a partir do vértice raiz foram visitados. Durante a busca em largura, uma lista de predecessores pode ser criada para armazenar os predecessores de cada vértice encontrado, permitindo que o caminho mais curto até um determinado vértice seja encontrado.


## 3. How many edges does a complete graph with n vertices have? What about a  complete directed graph with n vertices?

  Um grafo completo com n vértices tem n*(n-1)/2 arestas. Isso ocorre porque cada um dos n vértices está conectado a n-1 vértices (n-1 arestas), mas cada aresta é contada duas vezes, uma para cada vértice conectado a ela, então precisamos dividir por 2. Contudo, um grafo completo dirigido com n vértices possui n*(n-1) arestas. Isso ocorre porque cada vértice tem n-1 vértices de saída, então o número total de arestas é n*(n-1). Note que a direção das arestas faz diferença, portanto, o número de arestas é diferente do caso não direcionado.





## 4. What are isomorphic graphs? Draw an example.

Grafos isomórficos são grafos que possuem a mesma estrutura, ou seja, eles têm o mesmo número de vértices, as mesmas conexões e a mesma distribuição de graus, mas podem ter rótulos de vértices e arestas diferentes.







```python
import networkx as nx
import matplotlib.pyplot as plt


G1 = nx.Graph()
G1.add_edges_from([(1,2), (2,3), (3,1), (2,4)])
G2 = nx.Graph()
G2.add_edges_from([('A','B'), ('B','C'), ('C','A'), ('B','D')])


isomorphic = nx.is_isomorphic(G1, G2)


plt.subplot(121)
nx.draw(G1, with_labels=True)
plt.subplot(122)
nx.draw(G2, with_labels=True)
plt.suptitle('Grafos Isomórficos: {}'.format(isomorphic))
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Definição do grafo
V = [1, 2, 3, 4, 5, 6]
E = [(2, 5), (6, 1), (5, 3), (2, 3)]
G = nx.Graph()
G.add_nodes_from(V)
G.add_edges_from(E)

plt.subplot(121)
nx.draw(G, with_labels=True)
plt.title('Grafo Não Direcionado')

D = nx.DiGraph()
D.add_nodes_from(V)
D.add_edges_from(E)

plt.subplot(122)
nx.draw(D, with_labels=True)
plt.title('Grafo Dirigido')

plt.show()
```

## 8. How many edges does a graph have 3 vertices of degree 3 and one vertex of degree 5?

Um grafo com 3 vértices de grau 3 e um vértice de grau 5 tem um total de 12 arestas.

Para ver por que isso é verdade, observe que os três vértices de grau 3 estão todos conectados ao vértice de grau 5, portanto, este último vértice contribui com 5 arestas para o total. Além disso, cada um dos três vértices de grau 3 está conectado aos outros dois vértices de grau 3, portanto, cada um contribui com duas arestas adicionais. Portanto, o número total de arestas é dado por:

```python
5 + 2 + 2 + 2
```

```python
# Define the degrees of the vertices
deg = [3, 3, 3, 5]

# Calculate the total number of edges
num_edges = sum(deg) // 2

print("Total number of edges:", num_edges)
```

## 9. Mr. A is friend with Mrs. B, but she doesn't like him back. She has a reciprocal friendship with both C and D, but only C considers D a friend. D has also sent friend requests to E, F, G, and H but, so far, only G replied. G also has a reciprocal relationship with A. Draw the corresponding directed graph.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph object
G = nx.DiGraph()

# Add nodes to the graph
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])

# Add edges to the graph
G.add_edge('A', 'C')
G.add_edge('C', 'D')
G.add_edge('D', 'E')
G.add_edge('D', 'F')
G.add_edge('D', 'G')
G.add_edge('D', 'H')
G.add_edge('G', 'A')
G.add_edge('B', 'C')

# Draw the graph
pos = nx.circular_layout(G)
nx.draw_networkx(G, pos, with_labels=True)
plt.show()
```

## 10. Draw the graph from the previous exercise as undirected and weighted, with the weight being 2 if the connection is reciprocal, 1 otherwise.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()


G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])


G.add_edge('B', 'C', weight=1)
G.add_edge('B', 'D', weight=1)
G.add_edge('C', 'D', weight=2)
G.add_edge('C', 'E', weight=1)
G.add_edge('C', 'A', weight=2)
G.add_edge('A', 'G', weight=2)
G.add_edge('E', 'G', weight=1)
G.add_edge('D', 'F', weight=1)
G.add_edge('F', 'G', weight=1)


pos = nx.spring_layout(G)

# Obter pesos das arestas
weights = [G[u][v]['weight'] for u, v in G.edges()]

# Desenhar grafo com os pesos das arestas
nx.draw(G, pos, with_labels=True, font_size=12, node_color='lightblue', edge_color='gray', width=weights, font_weight='bold')

# Mostrar o grafo
plt.show()
```
